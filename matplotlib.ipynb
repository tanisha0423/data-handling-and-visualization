```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[4,5,6,7]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line graph")
plt.plot(x,y)
plt.show()
```

```python
x=[20,30,40,50]
plt.pie(x)
plt.show()
```

```python
import numpy as np
z=np.array(["Cherry","Mango","Watermelon","Apple"])
y=np.array(["blue","green","red","black"])
plt.pie(x,labels=z,colors=y)
plt.legend()
plt.show()
```

```python
pos=np.array([0,0.2,0,0])
plt.pie(x,labels=z,colors=y,explode=pos)
plt.legend()
plt.show()
```

```python
x=["A","B","c","D"]
y=[1,2,3,4]
plt.bar(x,y)
plt.show()
```

```python
x=["A","B","c","D"]
z=["blue","green","red","black"]
y=[1,2,3,4]
plt.barh(x,y,color=z)
plt.show()
```

```python
x=["A","B","c","D"]
size=[100,200,250,14]
y=[1,2,3,4]
z=["blue","green","blue","black"]
plt.scatter(x,y,color=z,marker='*',edgecolor="g",s=size,alpha=0.5)
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.show()
```
