```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    "Date":pd.to_datetime(["2022-01-01",'2022-02-02','2022-03-03','2022-04-02','2022-05-03','2022-07-09','2022-08-07','2022-10-12']),
    'Value':[10,12,22,24,29,30,35,37]
}
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df['Value'],marker='X')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time series with trend")
#plt.grid(True)
plt.show()
```
