```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
```

```python
text_data = "Python is a high-level programming language known for its simplicity, readability, and versatility. It emphasizes code readability and clear syntax, making it an ideal choice for beginners and experienced developers alike. Python supports multiple programming paradigms and has a vast ecosystem of libraries and frameworks, making it suitable for a wide range of applications, including web development, data analysis, artificial intelligence, automation, and more."
wordcloud = WordCloud(width=800,height=400,background_color='black').generate(text_data)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.show()
```

```python
words=text_data.split()
word_counts=Counter(words)

```

```python
plt.figure(figsize=(10,5))
plt.bar(word_counts.keys(),word_counts.values())
plt.title('Histogram')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```
