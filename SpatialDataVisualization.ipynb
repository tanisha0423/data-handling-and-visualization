```python
import folium

bangalore_locations = [
    {"Neighborhood": "Koramangala", "Latitude": 12.9358, "Longitude": 77.6141, "ImageURL": "https://imgmedia.lbb.in/media/2022/01/61efa39514a08c2400a7ce56_1643094933277.png", "Description": "Koramangala is a vibrant locality in Bangalore known for its lively atmosphere, bustling cafes, and shopping streets."},
    {"Neighborhood": "Indiranagar", "Latitude": 12.9784, "Longitude": 77.6408, "ImageURL": "https://i.ytimg.com/vi/B6Nf28_DMsg/maxresdefault.jpg", "Description": "Indiranagar is a posh residential and commercial area in Bangalore famous for its tree-lined streets, upscale restaurants, and nightlife."},
    {"Neighborhood": "Whitefield", "Latitude": 12.9698, "Longitude": 77.7499, "ImageURL": "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRlBhz9OthEJ8nhakWLSxRmojApX7RctEI12iqa-udymA&s", "Description": "Whitefield is a major IT hub in Bangalore with numerous tech parks, multinational companies, and residential complexes."},
    {"Neighborhood": "Jaynagar", "Latitude": 12.9293, "Longitude": 77.5828, "ImageURL": "https://i.ytimg.com/vi/nLQJqAp1hkw/maxresdefault.jpg", "Description": "Jaynagar is one of the oldest neighborhoods in Bangalore, known for its cultural heritage, traditional markets, and educational institutions."},
    {"Neighborhood": "Jakkur", "Latitude": 13.0780, "Longitude": 77.6065, "ImageURL": "https://imgs.mongabay.com/wp-content/uploads/sites/30/2021/04/05112800/jokicm-jakkur.jpg", "Description": "Jakkur is a locality in Bangalore known for its serene environment, Jakkur Aerodrome, and recreational activities like flying, boating, and picnics."}
]

mean_latitude = sum(loc['Latitude'] for loc in bangalore_locations) / len(bangalore_locations)
mean_longitude = sum(loc['Longitude'] for loc in bangalore_locations) / len(bangalore_locations)

bangalore_map = folium.Map(location=[mean_latitude, mean_longitude], zoom_start=12)

for loc in bangalore_locations:
    html = f'<div><h4>{loc["Neighborhood"]}</h4><img src="{loc["ImageURL"]}" alt="Image" width="200"><p>{loc["Description"]}</p></div>'
    folium.Marker(
        location=[loc['Latitude'], loc['Longitude']],
        popup=loc['Neighborhood'],
        tooltip=html,
        parse_html=True
    ).add_to(bangalore_map)

bangalore_map
```
